```python
import matplotlib.pyplot as plt
# data
x=[1,2,3,4,5]
y=[2,4,6,8,10]
#plot
plt.plot(x,y,color='blue',linestyle='--', marker=0)
```

```python
#bar chat
import matplotlib.pyplot as plt
#data
categories=['Apples','Bananas','cherries','dates']
values=[30,45,10,25]
#plot
plt.bar(categories,values,color='orange')
plt.title('bar chart example')
plt.xlabel('fruit type')
plt.ylabel('Quantity')
plt.grid(True)
plt.show()
```
